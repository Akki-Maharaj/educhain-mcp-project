```python
# Verify installation by importing EduChain
from educhain import Educhain
print("EduChain imported successfully.")
```

```python
# Initialize the EduChain client
client = Educhain()

# Generate 3 multiple-choice questions on the topic "Python Programming Basics"
mcqs = client.qna_engine.generate_questions(
    topic="Python Programming Basics",
    num=3,
    question_type="Multiple Choice"
)

# Print the generated questions as JSON
print(mcqs.model_dump_json())
```

```python
# Generate a lesson plan for "Python Programming Basics"
lesson_plan = client.content_engine.generate_lesson_plan(
    topic="Python Programming Basics"
)

# Print the lesson plan as JSON
print(lesson_plan.model_dump_json())
```
